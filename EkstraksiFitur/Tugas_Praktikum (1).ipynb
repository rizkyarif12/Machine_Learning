```python
# Langkah 1: Memuat Dataset
import pandas as pd

# Ganti 'file.csv' dengan path file yang sesuai
file_path = 'wbc.csv'
data = pd.read_csv(file_path)

# Langkah 2: Memisahkan Variabel yang Dapat dan Tidak Dapat Digunakan
# Kolom 'id' dan 'Unnamed: 32' tidak relevan, jadi kita akan menghapusnya
data_cleaned = data.drop(columns=['id', 'Unnamed: 32'])

# Menampilkan nama kolom setelah penghapusan
data_cleaned.columns
```

```python
# Langkah 3: Melakukan Encoding pada Kolom 'Diagnosis'
from sklearn.preprocessing import LabelEncoder

# Menggunakan LabelEncoder untuk mengubah 'M' menjadi 1 dan 'B' menjadi 0
le = LabelEncoder()
data_cleaned['diagnosis'] = le.fit_transform(data_cleaned['diagnosis'])

# Menampilkan beberapa baris untuk memastikan perubahan
data_cleaned[['diagnosis']].head(10)
```

```python
# Langkah 4: Melakukan Standarisasi pada Semua Kolom Numerik
# # Menampilkan beberapa baris setelah standarisasi
# data_cleaned.head()
from sklearn.preprocessing import StandardScaler

# Separate features and diagnosis
features = data_cleaned.drop(columns=['diagnosis'])

# Standardize the numerical columns
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Combine the scaled features with the diagnosis column
df_scaled = pd.DataFrame(features_scaled, columns=features.columns)
df_scaled['diagnosis'] = data_cleaned['diagnosis']

# Display the first few rows to confirm scaling
print(df_scaled.head())
```

```python
# Langkah 5: Melakukan Stratified Split Data
from sklearn.model_selection import train_test_split

# Memisahkan fitur (X) dan label (y)
X = data_cleaned.drop(columns=['diagnosis'])
y = data_cleaned['diagnosis']

# Melakukan stratified split dengan rasio 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Menampilkan ukuran data latih dan data uji
print(f'Data latih: {X_train.shape}, Data uji: {X_test.shape}')
```
